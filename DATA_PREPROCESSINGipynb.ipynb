```python
data2 = pd.read_csv('/content/drive/MyDrive/FEDFUNDS.csv')
data2
```

```python

data['Date'] = pd.to_datetime(data['Date'])
data1['Date'] = pd.to_datetime(data1['DATE'])
data['GDP'] = None
for index, row in data.iterrows():
    date = row['Date']
    quarter_gdp = data1.loc[data1['Date'] <= date].iloc[-1]['GDPC1']
    data.at[index, 'GDP'] = quarter_gdp
data
```

```python
data2
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data2['Date'] = pd.to_datetime(data2['DATE'])

data['FEDFUNDS'] = None
for index, row in data.iterrows():
    date = row['Date']
    month_gdp = data2.loc[data2['Date'].dt.to_period('M') == date.to_period('M')]['FEDFUNDS']
    if month_gdp.empty:
        month_gdp = data2.loc[data2['Date'] <= date].iloc[-1]['FEDFUNDS']
    else:
        month_gdp = month_gdp.iloc[0]
    data.at[index, 'FEDFUNDS'] = month_gdp
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data4['Date'] = pd.to_datetime(data4['TIME'])
data['Inflation'] = None

for index, row in data.iterrows():
    date = row['Date']
    month_gdp = data4.loc[data4['Date'].dt.to_period('M') == date.to_period('M')]['Value']
    if month_gdp.empty:
        month_gdp = data4.loc[data4['Date'] <= date].iloc[-1]['Value']
    else:
        month_gdp = month_gdp.iloc[0]
    data.at[index, 'Inflation'] = month_gdp

data
```
